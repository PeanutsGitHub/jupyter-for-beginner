# Working With JSON Data in Python

### Open JSON from Dictionary

```python
# Open JSON from Dictionary

import json

# Dictionary
json_dict = {
    "person": [{
        "name": "John",
        "age": 30,
        "married": True,
        "divorced": False,
        "children": ("Ann", "Billy"),
        "pets": None,
        "cars": [
                {"model": "BMW 230", "mpg": 27.5},
                {"model": "Ford Edge", "mpg": 24.1}
        ]
    },
        {
        "name": "Jack",
        "age": 35,
        "married": True,
        "divorced": False,
            "children": ("Anna", "Tom"),
            "pets": ("cat"),
            "cars": [
                {"model": "BMW 230", "mpg": 27.5}
            ]
    }
    ]}

print("json_dict :", type(json_dict))

# Load Dictionary to String
persons_str = json.dumps(json_dict, indent=2)
print("persons_str:", type(persons_str))

# Load String to JSON -> json.loads
persons = json.loads(persons_str)
print("persons:", type(persons))

# Print JSON All
print(persons)

# Print JSON by detail
for person in persons['person']:
    print(person['name'], person['age'])

# Print JSON to String with Indent -> json.dumps
print(json.dumps(persons, indent=2))

# Export JSON to JSON File -> json.dump
with open("person.json", "w") as f:
    json.dump(persons, f, indent=2)
```

### Open JSON from String (from JSON File Format)

```python
# Open JSON from String (from JSON File Format)

import json

# String from JSON File
json_str = """
{
    "person": [
        {
            "name": "John",
            "age": 30,
            "married": true,
            "divorced": false,
            "children": [
                "Ann",
                "Billy"
            ],
            "pets": null,
            "cars": [
                {
                    "model": "BMW 230",
                    "mpg": 27.5
                },
                {
                    "model": "Ford Edge",
                    "mpg": 24.1
                }
            ]
        },
        {
            "name": "Jack",
            "age": 35,
            "married": true,
            "divorced": false,
            "children": [
                "Anna",
                "Tom"
            ],
            "pets": "cat",
            "cars": [
                {
                    "model": "BMW 230",
                    "mpg": 27.5
                }
            ]
        }
    ]
}
"""


print("json_str :", type(json_str))

# Load Dictionary to String
# persons_str = json.dumps(json_dict, indent=2)
# print("persons_str:", type(persons_str))

# Load String to JSON -> json.loads
persons = json.loads(json_str)
print("persons:", type(persons))

# Print JSON All
print(persons)

# Print JSON by detail
for person in persons['person']:
    print(person['name'], person['age'])

# Print JSON to String with Indent -> json.dumps
print(json.dumps(persons, indent=2))

# Export JSON to JSON File -> json.dump
with open("person.json", "w") as f:
    json.dump(persons, f, indent=2)
```

### Open JSON from JSON File Format in Local

```python
# Open JSON from JSON File Format in Local

import json

# Load JSON File to JSON -> json.load
with open("person.json") as json_file:
    persons = json.load(json_file)
print("persons:", type(persons))

# Print JSON All
print(persons)

# Print JSON by detail
for person in persons['person']:
    print(person['name'], person['age'])

# Print JSON to String with Indent -> json.dumps
print(json.dumps(persons, indent=2))

# Export JSON to JSON File -> json.dump
with open("person.json", "w") as f:
    json.dump(persons, f, indent=2)
```

### Open JSON from JSON File Format in URL (Direct to JSON)

```python
# Open JSON from JSON File Format in URL (Direct to JSON)

import json
import requests

# Making a get request
response = requests.get(
    'https://raw.githubusercontent.com/PeanutsGitHub/data_files/main/person.json')
persons = response.json() # Load Direct to JSON
print("persons:", type(persons))

# Print JSON All
print(persons)

# Print JSON by detail
for person in persons['person']:
    print(person['name'], person['age'])

# Print JSON to String with Indent -> json.dumps
print(json.dumps(persons, indent=2))

# Export JSON to JSON File -> json.dump
with open("person.json", "w") as f:
    json.dump(persons, f, indent=2)
```

### Open JSON from JSON File Format in URL (Text to JSON)

```python
# Open JSON from JSON File Format in URL

import json
import requests

# Making a get request
response = requests.get('https://raw.githubusercontent.com/PeanutsGitHub/data_files/main/person.json')
json_file = response.text # Load to Text
print("json_file:", type(json_file))

# Load JSON File to JSON -> json.loads
persons = json.loads(json_file)
print("persons:", type(persons))

# Print JSON All
print(persons)

# Print JSON by detail
for person in persons['person']:
    print(person['name'], person['age'])

# Print JSON to String with Indent -> json.dumps
print(json.dumps(persons, indent=2))

# Export JSON to JSON File -> json.dump
with open("person.json", "w") as f:
    json.dump(persons, f, indent=2)
```

### Real world JSON with Sort, Filter and Unique

```python
# Open JSON from JSON File Format in URL (Direct to JSON)

import json
import requests
import numpy as np

# Original File: 'https://api.publicapis.org/entries'
# Backup File: 'https://raw.githubusercontent.com/PeanutsGitHub/data_files/main/api.publicapis.org_entries.json'

# Making a get request
response = requests.get(
    'https://raw.githubusercontent.com/PeanutsGitHub/data_files/main/api.publicapis.org_entries.json')
apis = response.json()  # Load Direct to JSON
print("apis:", type(apis))

"""
{
  "count": 1425,
  "entries": [
    {
      "API": "AdoptAPet",
      "Description": "Resource to help get pets adopted",
      "Auth": "apiKey",
      "HTTPS": true,
      "Cors": "yes",
      "Link": "https://www.adoptapet.com/public/apis/pet_list.html",
      "Category": "Animals"
    },
"""

# Print JSON to String with Indent -> json.dumps
#print(json.dumps(apis, indent=2))

# api = apis['entries']
# Print JSON to String with Indent -> json.dumps
# print(json.dumps(api, indent=2))

# Sort data in JSON
apis_sort = sorted(apis['entries'], key=lambda k: k['API'], reverse=False)
print("apis_sort:", json.dumps(apis_sort[:5], indent=2))  # Print Head 5


def list_unique(unique):  # Method List Unique Key and Count Value
    unique_list = [item[unique] for item in apis['entries']]
    keys, values = np.unique(unique_list, return_counts=True)
    return dict(zip(keys, values))


# Group of Unique Value and Count
auths = list_unique('Auth')
https = list_unique('HTTPS')
cors = list_unique('Cors')
categorys = list_unique('Category')

print("Auth: ", auths)
print("HTTPS: ", https)
print("Cors: ", cors)
print("Category: ", categorys)

# Filter data in JSON
lv = "Animals"
print("Category:", lv)
apis_category = [item for item in apis['entries']
                 if lv in item['Category']]
print(json.dumps(apis_category[:5], indent=2))  # Print Head 5
print("len(apis_category):", len(apis_category))

for api in apis_category:
    print(api['API'], ":", api['Description'])
```
