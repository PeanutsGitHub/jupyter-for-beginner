```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
import os
# Change the current working directory
# os.chdir(os.path.dirname(__file__))  # Python .py
os.chdir(os.path.abspath(''))  # Python & Python Notebook .ipynb
```

```python
data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
#Ref: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# <style> white, dark, whitegrid, darkgrid, ticks # Ex. 'seaborn-v0_8-whitegrid'
plt.style.use('seaborn-v0_8')

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
